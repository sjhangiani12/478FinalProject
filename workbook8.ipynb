```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('marijuana-use-2016.csv')
```

```python
df = df[df.Region == "United States"].reset_index(drop=True)
```

```python
df = df[df.Variable == "Marijuana Use"]
df
```

```python
first = df[df['Marijuana Use'] == "First Use of Marijuana"].reset_index(drop=True)
```

```python
import seaborn as sns
sns.lineplot(data=first, x='Year', y='Value', hue='Age Range')
```

```python
month = df[df['Marijuana Use'] == "Marijuana Use in the Past Month"].reset_index(drop=True)
```

```python
year = df[df['Marijuana Use'] == "Marijuana Use in the Past Year"].reset_index(drop=True)
```

### First Use of Marijuana Over Years

```python
import seaborn as sns
p = sns.lineplot(data=first, x='Year', y='Value', hue='Age Range')
```

### Marijuana Use in the Past Month

```python
import seaborn as sns
sns.lineplot(data=month, x='Year', y='Value', hue='Age Range')
```

### Marijuana Use in the Past Month

```python
import seaborn as sns
sns.lineplot(data=year, x='Year', y='Value', hue='Age Range')
```

The above charts show us quite a bit of interesting information about the overall trend of marijuana use and can help us understand where it's going in the future. First off, it is evident that Marijuana is a drug that is especially common in younger adults, seen by the 18-25 rate for all three metrics, first time users, user in the last month, and user in the last year being the highest. In addition, it shows us the popularity of the drug in young adults, and how much more popular it is at the age.
